```python
class Circle:
    def __init__(self, radius = 1.0, color = 'red'):
        # if not isinstance(radius, float):
        #     raise ValueError('SHOULD BE FLOAT')
        # if not isinstance(color,str):
        #     raise ValueError('SHOULD BE STRING')
        
        
        self.__radius = radius
        self.__color = color
        
    @property
    def radius(self): return self.__radius
    
    @radius.setter
    def radius(self, radius): 
        self.__radius = radius
    
    @property
    def color(self): return self.__color
    
    @color.setter
    def color(self, color):
        self.__color = color 
        
    def getArea(self): return self.__radius * 3.14
    
    def __str__(self):
        return f'Circle[radius = {self.__radius}, color ={self.__color}]'
    
    
        
class Cylinder(Circle):
    def __init__(self, radius=1, color='red',height = 1.0):
        super().__init__(radius, color)
        self.__height = height
        
    @property
    def height(self):
        return self.__height
    
    @height.setter
    def height(self, height):
        self.__height = height
        
    '''surface Area = 2pir^2 + 2pirh'''
    def getArea(self):
        return 2 * 3.14 * (self.__radius)**2 + 2* 3.14 * self.__radius * self.__height
    def getVolume(self):
        return self.getArea * self.__height
    
    
def testCylinder():
    CylinderList = [
        Cylinder(),
        Cylinder(10, 0),
        Cylinder(2.0, 10.0)
    ]
    for cylinder in CylinderList:
        print(f'Radius: {cylinder.radius} | Color: {cylinder.color} | Height: {cylinder.height}')
        
        
testCylinder()
```

a. Xây dựng lớp hình chữ nhật (Rectangle) bao gồm: 
- Length, width 
- Constructors 0 tham số và 2 tham số: nếu length và width là số âm thì mặc định 
bằng 0 
- Getters, setters: nếu length và width là số âm thì mặc định bằng 0 
- Hàm tính diện tích getarea(): protected 
- Hàm tính chu vi getperimeter(): protected 
- Override hàm tostring(): return chuỗi thông tin (length, width)

```python
class Rectangle:
    def __init__(self, length, width):
        self.__length = 0 if length < 0 else length
        self.__width = 0 if width < 0 else width
        
    @property
    def length(self):
        return self.__length
    
    
    @length.setter
    def length(self, value):
        self.__length = 0 if value <= 0 else  value
        
        
    @property
    def width(self):
        return self.__width
    
    @width.setter
    def width(self, value):
        self.__width = 0 if value <=0 else value
        
        
    def _getArea(self):
        return self.__length * self.__width
    
    def _getPerimeter(self):
        return (self.__length + self.__width) * 2
    
        
    def __str__(self) -> str:
        return f'Rectangle[length = {self.__length}, width = {self.__width}]'
        
    
    
class Box(Rectangle):
    def __init__(self, length, width, height):
        super().__init__(length, width)
        self.__height = height
        if self.__height | self.__height ==0:
            self.__height = 0 if height <= 0 else height
            
            
    @property
    def height(self):
        return self.__height
    
    @height.setter
    def height(self, value):
        self.__height = 0 if value <= 0 else value
        
        
    def __str__(self):
        print(f"Height': {self.__height}")
        return super().__str__()
        
        
    def getVolume(self):
        return super()._getArea() * self.__height
    
    
    
class TestBox:
    # def __init__(self):
    #     print("Create testbox")
    @staticmethod
    def main():
        rec1 = Rectangle(10, 3)
        box1 = Box(2, 4, 6)
        print(rec1._getArea())
        print(rec1._getPerimeter())
        
        print(box1._getArea())
        print(box1.getVolume())
        
        
test = TestBox()
test.main()
# sai về sử dụng len(self.__height)
# sai quên print() ở main
# sử dụng staticmethod hoặc cho self vào main
        
```
