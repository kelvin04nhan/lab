```python
from abc import ABCMeta, abstractmethod 
```

```python

class Document(metaclass = ABCMeta):
    def __init__(self, fileName = 'Document1', createdDate='now', lastModifiedDate ='null', content=''):
        self.fileName = fileName
        self.createDate = createdDate
        self.lastModifiedDate = lastModifiedDate
        self.content = content
    @abstractmethod
    def inputContent(self):
        return NotImplemented
    
    @abstractmethod
    def save(self):
        return NotImplemented
    
    @abstractmethod
    def saveAs(self, newName):
        return NotImplemented
    
    @abstractmethod
    def open(self):
        return NotImplemented

```

```python

class TextPad(Document):
    def __init__(self,words = 0,size =0, fileName='NoName.txt', createdDate='now', lastModifiedDate='null', content=''):
        super().__init__(fileName , createdDate, lastModifiedDate, content)
        self.words = words
        self.size = size
    
    def getWords(self):
        pass
    
    def getSize(self):
        return len(self.content)
    
    def inputContent(self):
        self.content = input("Enter your word: ")
        
    def save(self):
        self.words = self.getWords()
        self.size = self.getSize()
        self.lastModifiedDate = 'now'
        
    def saveAs(self, newName):
        self.fileName = newName
        self.words = self.getWords()
        self.size = self.getSize()
        self.lastModifiedDate = "now"

    def open(self):
        infomation = "{}\n{}\nCreated date: {}\nLast modified date: {}\nWords: {}\nSize: {}".format(self.fileName, self.content, self.createdDate, self.lastModifiedDate, self.words, self.size)
        print(infomation)
        
```

```python
def createNewTextPad():
    global textpad, folder
    textpad = TextPad()
    folder.append(textpad)
    
def inputTextPad():
    global textpad
    textpad.inputContent()

def saveTextPad():
    global textpad
    textpad.save()
    print("Saved")

def saveAsTextPad():
    global textpad
    newName = input("Enter new name for TextPad")
    textpad.saveAs(newName)
    print("Saved as {}".format(newName))

def openTextPad():
    global textpad
    textpad.open()
    
def exit():
    global state
    state = "off"
def printMenu(options):
    for optionNum, Option in options.items():
        print("[{}]. {}".format(optionNum, Option[0]))

options = {
    1: ["New TextPad", createNewTextPad],
    2: ["Input TextPad", inputTextPad],
    3: ["Save TextPad", saveTextPad],
    4: ["Save as TextPad", saveAsTextPad],
    5: ["Open TextPad", openTextPad],
    6: ["Exit", exit]
}

folder = []
state = "on"
createNewTextPad()
printMenu(options)
while state == "on":
    optionNum = int(input("Enter option number: "))
    method = options[optionNum][1]
    method()
for textpad in folder:
    print(textpad)
```
